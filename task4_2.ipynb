```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def addzeros(x):
    n = len(x)
    n = n - 1
    n = n | (n >> 1)
    n = n | (n >> 2)
    n = n | (n >> 4)
    n = n | (n >> 8)
    n = n | (n >> 16)
    n = n + 1
    x = np.concatenate((x,np.zeros((n-len(x)),dtype=complex)))
    return x
```

```python
def fft(x):
    N = len(x)
    if N == 1:
        return x
    else:
        X_even = fft(x[::2])
        X_odd = fft(x[1::2])
        X = [0] * N
        for k in xrange(N / 2):
            twiddle = complex()
            twiddle = np.exp(2j * np.pi * k / N)
            X[k] = X_even[k] + twiddle * X_odd[k]
            X[k + N / 2] = X_even[k] - twiddle * X_odd[k]
    return X
```

```python
def ifft(x):
    time_signal = fft([elem.conjugate() for elem in x])
    return [elem.conjugate() / len(x) for elem in time_signal]
```

```python
x = np.linspace(0,10,100)
```

```python
def f(x):
    return x

g = fft(addzeros(f(x)))
D = 1
for i in range(len(g) - D,len(g)):
    g[i] = 0
f2 = ifft(g)
f2 = np.array(f2).real

plt.plot(x,f2[:len(x):1],x, f(x))
plt.title("y = x")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
```
